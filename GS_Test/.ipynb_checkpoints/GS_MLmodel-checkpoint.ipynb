```python
# Initial imports
from path import Path
from pathlib import Path
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
# import CSV file into dataframe
file_path = Path("GS_SentScoreswVOL.csv")
df = pd.read_csv((file_path), index_col='Date')
df.index = pd.to_datetime(df.index)
df.index.sort_values()
df.dropna()
```

```python
X_train= df.iloc[0:12,4:8].copy()
X_train.dropna()
```

```python
X_test= df.iloc[13:18,4:8].copy()
X_test.dropna()
```

```python
y_train= df.iloc[0:12,8:9].copy()
y_train.dropna()
```

```python
y_test=df.iloc[13:19,8:9].copy()
y_test
```

```python
scaler = StandardScaler()

# Fitting Standard Scaler
X_scaler = scaler.fit(X_train)

# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_train_scaled
```

```python
# Create the Linear Regression model
model = LinearRegression()

# Fit the model to the data (Train the model)
model.fit(X_train_scaled, y_train)

# Predict the temperature using the week of the year as inputs
predictions = model.predict(X_train_scaled)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# The default scoring method is the R2 score
score = model.score(X_train_scaled, y_train)
r2 = r2_score(y_train, predictions)

print(f"Score: {score}, r2: {r2}")
```
